```python
from sklearn.model_selection import train_test_split
xtr,xts,ytr,yts=train_test_split(ip,op,test_size=0.2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, TfidfTransformer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
```

```python
model = Pipeline([('vector',TfidfVectorizer()),
                  ('tfidf',TfidfTransformer()),
                  ('algo',MultinomialNB())])
```

```python
model.fit(xtr.text,ytr.label) # here i can only feed 1 input column , how to put multiple columns
```

```python
pred1=model.predict(xts.text)
```

```python
pred1
```

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score

ac=accuracy_score(pred1,yts.label)
re=recall_score(pred1,yts.label)
pr=precision_score(pred1,yts.label)
f1=f1_score(pred1,yts.label)

print('Accuracy:',ac)
print('Recall:',re)
print('Precision:',pr)
print('F1 Score:',f1)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm=confusion_matrix(pred1,yts.label)
cm
```
